## Combine attributes into sequences

Correlation among data points seems to be weak. It would be useful to atempt to collect more temporal data to identify patterns over time.

Collapsing data into sequences associated with a specific sender might be an approach. This will mean that the models that work with time series data will be necessary to classify the value. One of such approaches is long-short term memory LSTM network (a type of recurrent neural network RNN).

The goal is to aggregate data into:
```
data = [
[
    [sendTime1, sendTime2, ..., sendTime15],
    [posx1, posx2, ..., posx15],
    [posy1, posy2, ..., posy15],
    [spdx1, spdx2, ..., spdx15],
    [spdy1, spdy2, ..., spdy15],
    [aclx1, aclx2, ..., aclx15],
    [acly1, acly2, ..., acly15],
    [hedx1, hedx2, ..., hedx15],
    [hedy1, hedy2, ..., hedy15],
],
...
]

labels = [
  0, 1, 0, 1 ...
]
```

I am intentionally skipping concrete class for now to simplify attack detection. The reason is that there might be different classes of attack in one sequence.

### Normalize data

Normalize data into ranges from 0 to 1. It will make sure large numbers do not influence the model.

````python
# order before working on it
df.sort_values(['senderPseudo', 'sendTime'], inplace=True)

# TODO: normalize data

FEATURE_COLS = ['sendTime', 'posx', 'posy', 'spdx', 'spdy', 'aclx', 'acly', 'hedx', 'hedy']

# collect sender events into batches and label if it is under attack
batches, batchLabels = list(), list()

for sender, senderGroup in df.groupby('senderPseudo'):
    window_size = 15
    overlap = 5
    if window_size > len(senderGroup):
        continue

    for i in range(0, len(senderGroup) - window_size + 1, window_size - overlap):
        # select items in window range
        senderWindowGroup = senderGroup.iloc[i: i + window_size]
        # if at least one event is considered an attack then label the whole series as such
        label = 1 if senderWindowGroup['stateInt'].max() == 3 else 0

        # choose only data which is interesting
        usefulValues = senderWindowGroup[FEATURE_COLS]
        # convert colmns to rows to feed as multivariate feature detection
        usefulValues = usefulValues.transpose()
        # add values
        batches.append(usefulValues.values)
        batchLabels.append(label)

npBatches, npBatchLabels = np.array(batches), np.array(batchLabels)
print(npBatches.shape, npBatchLabels.shape)
````

## Randomize and split test data

Randomize all series to be able to split data into train and test.

````python
import math

randomizedIndexes = np.random.permutation(len(npBatches))
x_all = npBatches[randomizedIndexes]
y_all = npBatchLabels[randomizedIndexes]

splitIndex =  math.ceil(len(x_all) * 0.8)

x_train, x_test = x_all[:splitIndex], x_all[splitIndex:]
y_train, y_test = y_all[:splitIndex], y_all[splitIndex:]


print(x_train.shape, "first 0 feature", x_train[0][0][0], "last 0 feature", x_train[-1][-1][0])
print(x_test.shape, "first 0 feature", x_test[0][0][0], "last 0 feature", x_test[-1][-1][0])
print(y_train.shape, "first label", y_train[0], "last label", y_train[-1])
print(y_test.shape, "first label", y_test[0], "last label", y_test[-1])
````

````python
import keras

LSTM_SIZE = 128
DROPOUT = 0.2
MID_SIZE = 0
MID_ACT = "relu"

inputs = keras.layers.Input(shape=(x_train.shape[1], x_train.shape[2]))

# first layer is LSTM
# dropout is used to avoid overfitting
lastOutput = keras.layers.LSTM(LSTM_SIZE, dropout=DROPOUT)(inputs)

# intermediate layer
if MID_SIZE > 0:
    lastOutput = keras.layers.Dense(MID_SIZE, activation=MID_ACT)(lastOutput)

# output layer
# sigmoid will compress the value into 0-1 range
outputs = keras.layers.Dense(1, activation="sigmoid")(lastOutput)

model = keras.Model(inputs=inputs, outputs=outputs)
# binary_crossentropy is used as it is a binary classification problem
model.compile(optimizer=keras.optimizers.Adam(learning_rate=0.001), loss="binary_crossentropy")
model.summary()
````

````python
path_checkpoint = "model_checkpoint.weights.h5"
es_callback = keras.callbacks.EarlyStopping(monitor="val_loss", min_delta=0, patience=5)

modelckpt_callback = keras.callbacks.ModelCheckpoint(
    monitor="val_loss",
    filepath=path_checkpoint,
    verbose=1,
    save_weights_only=True,
    save_best_only=True,
)

batch_size = 200 # how many samples to use for one test iteration
epochs = 100 # how many times to go through the whole dataset to try and reduce the loss

history = model.fit(
    x_train, 
    y_train,
    batch_size=batch_size, 
    epochs=epochs,
    validation_data=((x_test, y_test)),
    callbacks=[es_callback, modelckpt_callback],
)
````

````python
def visualize_loss(history, title, batch_size=0, LSTM_SIZE=0, DROPOUT=0, MID_SIZE=0, MID_ACT="", normalized=False):
    loss = history.history["loss"]
    val_loss = history.history["val_loss"]
    epochs = range(len(loss))
    plt.figure()
    plt.plot(epochs, loss, "b", label="Training loss")
    plt.plot(epochs, val_loss, "r", label="Validation loss")
    plt.title(title)
    plt.xlabel("Epochs")
    plt.ylabel("Loss")
    plt.legend()
    plt.axis([0, len(loss), 0, 1])
    plt.text(2, 0.95, 
             f"Batch={batch_size}\nLSTM={LSTM_SIZE}\nLSTM dropout={DROPOUT}\nMid size={MID_SIZE}\nMid activation={MID_ACT}\nNormalized={normalized}", 
             fontsize=10, 
             wrap=True,
             verticalalignment='top', 
             bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))
    plt.show()


visualize_loss(history, "Training and Validation Loss", batch_size, LSTM_SIZE, DROPOUT, MID_SIZE, MID_ACT)
````

# Evaluate the model
Use confusion matrix to assess accuracy, precission, reclaa and F1 score

````python
y_prediction = model.predict(x_test)
y_prediction
````

````python
# round the predictions as those are probabilities
np.around(y_prediction,0,y_prediction)

print(y_test.shape)
print(y_prediction.shape)

print("True normal events", np.count_nonzero(y_test==0))
print("True attack events", np.count_nonzero(y_test==1))
print("Predicted normal events", np.count_nonzero(y_prediction==0))
print("Predicted attack events", np.count_nonzero(y_prediction==1))
````

````python
from sklearn.metrics import ConfusionMatrixDisplay, confusion_matrix
result = confusion_matrix(y_test, y_prediction)
disp = ConfusionMatrixDisplay(confusion_matrix=result, display_labels=['Normal', 'Attack'])
disp.plot()
plt.show()
````

````python
from sklearn.metrics import precision_score, accuracy_score, recall_score, f1_score

# (true positive + true negative) / all
accuracy = accuracy_score(y_test, y_prediction)
print("Accuracy", accuracy)

# Precision measures the model trustiness in classifying positive (is attack) samples
# true positive / (true positive + true negative)
precision = precision_score(y_test, y_prediction, pos_label=1)
print("Precision", precision)

# Recall measures how many positive samples were correctly classified by the model
# true positive / (true positive + false negative)
recall = recall_score(y_test, y_prediction, pos_label=1)
print("Recall", recall)

# The F1 score can be interpreted as a harmonic mean of the precision and recall
f1 = f1_score(y_test, y_prediction, pos_label=1)
print("F1 score", f1)

metricLabels = ['Accuracy','Precision','Recall','F1 score'] 
metricValues = [accuracy, precision, recall, f1] 
  
X_axis = np.arange(len(metricLabels)) 
plt.bar(X_axis, metricValues, 0.4)
plt.xticks(X_axis, metricLabels) 
plt.xlabel("Metrics") 
plt.title("Model evaluation scores") 
plt.show() 
````
